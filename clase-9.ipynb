```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df= pd.read_csv('../input/airlinedata/airline_data.csv',
               low_memory=False)
df.head()
```

```python
data = df.sample (n = 500,
                  random_state = 42)
data.shape
```

```python
data = df.head(500)
```

```python
fig = go.Figure(data = go.Scatter ( x = data ['Distance'],
                                   y = data ['DepTime'],
                                   mode = 'markers',
                                   marker = dict (color = 'red')
                                  )
               )
fig.show()
```

```python
bar_data = data.groupby (['DestState'])['Flights'].sum().reset_index()
bar_data.head()
```

```python
fig = px.bar(bar_data,
            x = 'DestState',
            y = 'Flights')
fig.show()
```
