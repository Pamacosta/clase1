## 1. Existe alguna relaci√≥n entre Head Size(cm^3) vs Brain Weight(grams)

```python
df= pd.read_csv('headbrain.csv')
df.head()

x = df['Head Size(cm^3)']
y = df['Brain Weight(grams)']

np.corrcoef (x,y)
```

```python
colors = {1: 'red', 2: 'green'}
plt.figure(figsize = (13,10))
plt.scatter(x,y,
            c= df['Gender'].map(colors),
            marker = 'x',
            s = 175
           )
plt.xlabel('Head Size(cm^3)', fontweight = 'regular', color = 'black',
fontsize = 18, horizontalalignment = 'center')
plt.ylabel('Brain Weight(grams)',fontweight = 'regular', color = 'black',
fontsize = 18, horizontalalignment = 'center')
```

#### Conclusi√≥n:
La coeficiente de correlaci√≥n es 0.7995. Se puede decir que existe una correlacci√≥nmuy fuerte entre las variabels head size vs Brain Weight
