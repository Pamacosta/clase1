```python
#nnp.arange (start, end, step)

x = np.arange(0, 3*np.pi, 1)
y = np.sin(x)
plt.plot(x,y)
```

```python
ysin = np.sin(x)
ycos = np.cos(x)

plt.plot(x,ysin)
plt.plot(x,ycos)

plt.xlabel("X axis label")
plt.ylabel("Y axis label")

plt.title('sine & cosine funtions')
```

```python
plt.subplot(2,2,2) #2 rows, 1 columns, postition 1 

plt.plot(x,ysin)
plt.title('Sine')
        
plt.subplot(2,2,1) #2 rows, 1 columns, postition 2
plt.plot(x,ycos)
plt.title('Cosine')

plt.subplot(2,2,3) #2 rows, 1 columns, postition 1 

plt.plot(x,ysin)
plt.title('Sine')
        
plt.subplot(2,2,4) #2 rows, 1 columns, postition 2
plt.plot(x,ycos)
plt.title('Cosine')
```

```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head(2)
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'Height',
           data = df,
           fit_reg = False,
           hue = 'Gender'
          )
plt.title('Scatter plot of Wingspan vs Heigth')
```

```python
sns.swarmplot(x='Gender',
              y= 'Wingspan',
              data = df,
             )
plt.title('Gender vs Wingspan, the Swarmplot')
plt.show()
```
