```python
df = pd.read_excel ('datasets/Canada.xlsx',
                    sheet_name = 'Canada by Citizenship',
                    skiprows = range (20),
                    skipfooter = 2,
                    engine = 'openpyxl'
                   )
df.head(2)
```

```python
df.columns
```

```python
df.drop (['AREA','REG','DEV','Type','Coverage','Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],
         axis = 1,
         inplace = True)
```

```python
df.rename(columns ={'OdName': 'Country',
                   'AreaName': 'Continent',
                   'RegName': 'Region'},
         inplace= True)
```

```python
df.columns
```

```python
df['Total'] = df.iloc[:,4:].sum(axis = 1)
df.head(2)
```

```python
df.columns = list(map(str,
                     df.columns
                     )
                 )
#df.columns
fav_year = list(map(str,
                   range(1980,2014)
                   )
               )
#df.head(2)
df.set_index('Country',
            inplace = True)
```

```python
df.index.name = None
df.head(2)
```

```python
print (plt.style.available)
plt.style.use('ggplot')
```

```python
haiti = df.loc['Haiti', fav_year]
```

```python
haiti
```

```python
haiti.plot()
```

```python
dfCI = df.loc[['China', 'India'],
             fav_year]
dfCI = dfCI.transpose()

dfCI.plot()
plt.title('Immigration to Canada from India and China')
plt.ylabel('Immigration form China and India')
plt.xlabel('Years')
```

```python
df_top = df.sort_values(by = 'Total',
                       ascending = False,
                       axis = 0,
                       )
df_top = df_top.head(7)
```

```python
df_top.drop (['Total', 'Unknown'],
         axis = 0,
         inplace = True)
```

```python
df_top = df_top[fav_year]
df_top= df_top.transpose()
df_top.plot()
```
